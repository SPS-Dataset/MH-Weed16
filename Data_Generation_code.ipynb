```python
import os
import cv2

# Function to resize and convert image to PNG if necessary
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        # Resize image to 512x512
        img = cv2.resize(img, (512, 512))
        cv2.imwrite(image_path, img)
        return image_path
    return None

# Function to crop and save images based on YOLO annotations
def crop_images_with_yolo(image_folder, yolo_folder, output_folder):
    for txt_file in os.listdir(yolo_folder):
        if txt_file.endswith(".txt"):
            # Extract image name from YOLO txt file
            image_name = os.path.splitext(txt_file)[0]
            image_name = image_name + ".png"
            image_path = os.path.join(image_folder, image_name)
            if not os.path.isfile(image_path):
                continue
            image_path = preprocess_image(image_path)
            if image_path is None:
                continue
            
            img = cv2.imread(image_path)  # Load the image here

            with open(os.path.join(yolo_folder, txt_file), 'r') as f:
                lines = f.readlines()
                for line in lines:
                    parts = line.strip().split()
                    class_name = parts[0]
                    x_center, y_center, width, height = map(float, parts[1:])
                    # Calculate bounding box coordinates
                    left = int((x_center - width / 2) * 512)
                    top = int((y_center - height / 2) * 512)
                    right = int((x_center + width / 2) * 512)
                    bottom = int((y_center + height / 2) * 512)
                    # Ensure bounding box coordinates are within image bounds
                    left = max(0, left)
                    top = max(0, top)
                    right = min(512, right)
                    bottom = min(512, bottom)
                    # Crop the image if bounding box coordinates are valid
                    if right > left and bottom > top:
                        cropped_img = img[top:bottom, left:right]
                        # Save the cropped image if not empty
                        if cropped_img.size != 0:
                            output_filename = f"{image_name}_class_{class_name}.png"
                            cv2.imwrite(os.path.join(output_folder, output_filename), cropped_img)
                        else:
                            print(f"Empty cropped image: {output_filename}")

# Example usage
if __name__ == "__main__":
    image_folder = "D:/Josh AI Solution Dataset/soybean_defoliation/soybean_defoliation_with_weed/soybean_defoliation_with_weed_(13990_images)/"
    yolo_folder = "D:/Josh AI Solution Dataset/soybean_defoliation/soybean_defoliation_with_weed/soybean_defoliation_with_weed_(13990_images)/YOLO_darknet/"
    output_folder = "D:/JOSH_cropped_images_13990"
    
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    crop_images_with_yolo(image_folder, yolo_folder, output_folder)
```
